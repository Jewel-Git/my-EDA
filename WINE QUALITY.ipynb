## Visualizations

```python
data['quality'].value_counts().plot.bar()
```

*Most of the red wine recorded in the dataset have a quality rating between 5 and 6. Only a handful of the dataset records red wine with quality rated 8. Less number of red wines have a low quality rating (between 3 and 4).*

```python
data.hist(bins=10, figsize=(30, 25))
plt.show()
```

```python
columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
for col in columns:
    sns.distplot(data[col], bins=10, kde=True)
    plt.show()
```

*Only the PH and Density have a somewhat perfect normal distribution. The remaining variables are mostly left skewed except the target variable(Quality), volatile acidity and citric acid variables.*

```python
corrMatrix = data.corr()
corrMatrix
```

```python
plt.subplots(figsize=(20, 15))
sns.heatmap(corrMatrix, annot = True)
plt.show()
```

*Alcohol content has a somewhat high positive correlation with the quality of the red wines followed by sulphates content, while the volatile acidity has a strong negative correlation with the quality of the red wines.*
